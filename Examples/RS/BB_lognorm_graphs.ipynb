```python
import pickle
import pandas as pd
from matplotlib import pylab as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_pickle('BBresults.pkl')
pd.set_option('display.max_columns', None)

#creating distance columns
split_df = pd.DataFrame(df['distance'].tolist(), columns=['dist_meters', 'dist_degrees'])
df = pd.concat([df, split_df], axis=1)
dfB=df[df['channel'] == 'BHZ'] #sorting results for just broadband responses 
dfR=df[df['channel'] == 'EHZ']
dfRS=dfR[dfR['npts']==36001]
```

```python
dfRS
```

```python
#norm vs. depth
plt.figure()
dfB.plot(x ='edepth', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Ebbcorrpertaclognorm vs. Depth')
plt.ylim([-2,2])
plt.show()

plt.figure()
dfRS.plot(x ='edepth', y='ebbcorrpertaclognorm', kind = 'scatter', color='red')
plt.title('Ebbcorrpertaclognorm vs. Depth')
plt.ylim([-2,2])
plt.show()

#ax = dfB.plot(x='edepth', y='ebbcorrpertaclognorm', kind='scatter', color='blue')
#dfRS.plot(ax=ax, x='edepth', y='ebbcorrpertaclognorm', kind='scatter', color='red')
```

```python
#norm vs distance
plt.figure()
dfB.plot(x ='dist_degrees', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Broadband Ebbcorrpertaclognorm vs. Distance')
plt.ylim([-2,2])
plt.show()

plt.figure()
dfRS.plot(x ='dist_degrees', y='ebbcorrpertaclognorm', kind = 'scatter', color='red') 
plt.title('Raspberry Shake Ebbcorrpertaclognorm vs. Distance')
plt.ylim([-2,2])
plt.show()
```

```python
plt.figure()
dfB.plot(x ='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Ebbcorrpertaclognorm vs. Rake')
plt.ylim([-2,2])
plt.show()

plt.figure()
dfRS.plot(x ='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='red')
plt.title('Ebbcorrpertaclognorm vs. Rake')
plt.ylim([-2,2])
plt.show()
```

```python
#norm for strike slip 

df_strikeslip1=dfB[dfB['rake'] <-135] 
df_strikeslip2=dfB[dfB['rake'] >135]
df3=dfB[dfB['rake'] >-45]
df_strikeslip3=df3[df3['rake'] <45]

df_strikeslip1=df_strikeslip1.append([df_strikeslip2,df_strikeslip3])

plt.figure()
df_strikeslip1.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Strike-Slip Faulting')
plt.ylim([-2,2])
plt.show()

df_strikeslip11=dfRS[dfRS['rake'] <-135] 
df_strikeslip21=dfRS[dfRS['rake'] >135]
df31=dfRS[dfRS['rake'] >-45]
df_strikeslip31=df31[df31['rake'] <45]

df_strikeslip11=df_strikeslip11.append([df_strikeslip21,df_strikeslip31])

plt.figure()
df_strikeslip11.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='red')
plt.title('Strike-Slip Faulting')
plt.ylim([-2,2])
plt.show()
```

```python
#norm for normal faulting
df1=dfB[dfB['rake'] >-135]
dfB_normal=df1[df1['rake'] <-45]

plt.figure()
dfB_normal.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Normal Faulting')
plt.ylim([-2,2])

plt.show()

df11=dfRS[dfRS['rake'] >-135]
dfRS_normal=df11[df11['rake'] <-45]

plt.figure()
dfRS_normal.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='red')
plt.title('Normal Faulting')
plt.ylim([-2,2])

plt.show()
```

```python
#norm for thrust 
df2=dfB[dfB['rake'] >45]
df_thrust=df2[df2['rake'] <135]

plt.figure()
df_thrust.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='blue')
plt.title('Thrust Faulting')
plt.ylim([-2,2])

plt.show()

df21=dfRS[dfRS['rake'] >45]
dfRS_thrust=df21[df21['rake'] <135]

plt.figure()
dfRS_thrust.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter', color='red')
plt.title('Thrust Faulting')
plt.ylim([-2,2])

plt.show()
```

```python
#Example event analysis by distance
first_event=dfB[dfB['eventname'] == '2022071200']
lognorm = first_event['ebbcorrpertaclognorm'].tolist()
lognormmean=np.mean(lognorm) ###not sure if absolute value should be here 
print(lognormmean)

first_eventRS=dfRS[dfRS['eventname'] == '2022071200']
lognormRS = first_eventRS['ebbcorrpertaclognorm'].tolist()
lognormmeanRS=np.mean(np.abs(lognormRS))
print(lognormmeanRS)

ax = first_eventRS.plot(x='dist_degrees', y='ebbcorrpertaclognorm', kind='scatter', color='red')

plt.axhline(y=lognormmean)

#finding distance from BBmean

np.std(first_eventRS, axis=lognormmean)
```
