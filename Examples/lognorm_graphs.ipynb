```python
import pickle
import pandas as pd
from matplotlib import pylab as plt
%matplotlib inline
import numpy as np
```

```python
notdf = pd.read_pickle('Results3.pkl')
pd.set_option('display.max_columns', None)


df=notdf[notdf['npts'] ==36001] 
df
```

```python
df[df["ebbcorrpertaclognorm"]> 10].head()
```

```python
#norm vs. depth
plt.figure()
df.plot(x ='edepth', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Ebbcorrpertaclognorm vs. Depth')
plt.ylim([-2,2])
plt.show()
```

```python
#norm vs distance
plt.figure()
df.plot(x ='dist_degrees', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Ebbcorrpertaclognorm vs. Distance')
plt.ylim([-2,2])
plt.show()
```

```python
plt.figure()
df.plot(x ='rake', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Ebbcorrpertaclognorm vs. Rake')
plt.ylim([-2,2])
plt.show()
```

```python
#norm for strike slip 

df_strikeslip1=df[df['rake'] <-135] 
df_strikeslip2=df[df['rake'] >135]
df3=df[df['rake'] >-45]
df_strikeslip3=df3[df3['rake'] <45]

df_strikeslip1=df_strikeslip1.append([df_strikeslip2,df_strikeslip3])

plt.figure()
df_strikeslip1.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Strike-Slip Faulting')
plt.ylim([-2,2])
plt.show()
```

```python
#norm for normal faulting
df1=df[df['rake'] >-135]
df_normal=df1[df1['rake'] <-45]

plt.figure()
df_normal.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Normal Faulting')
plt.ylim([-2,2])

plt.show()
```

```python
#norm for thrust 
df2=df[df['rake'] >45]
df_thrust=df2[df2['rake'] <135]

plt.figure()
df_thrust.plot(x='rake', y='ebbcorrpertaclognorm', kind = 'scatter')
plt.title('Thrust Faulting')
plt.ylim([-2,2])

plt.show()
```
