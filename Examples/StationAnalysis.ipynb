```python
import pickle
import numpy as np
import pandas as pd 
```

```python
df=pd.read_pickle('NormMeanErgs.pkl')
df
```

```python
stationlist=df.drop_duplicates(subset=['Station Name'])
stationlist=stationlist['Station Name']

for station in stationlist:
    
    dfstation=pd.DataFrame
    dfstation=df[df['Station Name']==station]
    posdfstation=dfstation[dfstation['Normalization']>0]
    negdfstation=dfstation[dfstation['Normalization']<0]
    #posmean=np.mean(posdfstation['Normalization'])
    #negmean=np.mean(negdfstation['Normalization'])
    posdfstation['positive']=np.mean(posdfstation['Normalization'])
    negdfstation['negative']=np.mean(negdfstation['Normalization'])
    
    df_pos=pd.DataFrame()
    df_neg=pd.DataFrame()
    
    df_pos=pd.concat([df_pos,posdfstation],ignore_index=True)
    df_pos.to_pickle("RSstationsP.pkl")
    
    df_neg=pd.concat([df_neg,negdfstation],ignore_index=True)
    df_neg.to_pickle("RSstationsN.pkl")


    
    
    
```
