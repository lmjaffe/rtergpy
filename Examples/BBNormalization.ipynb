```python
import pickle
import pandas as pd
import numpy as np
import plotly.express as px
from matplotlib import pylab as plt
%matplotlib inline
import plotly.express as px
```

```python
df = pd.read_pickle('BBresults.pkl')
pd.set_option('display.max_columns', None)

#creating distance columns
split_df = pd.DataFrame(df['distance'].tolist(), columns=['dist_meters', 'dist_degrees'])
df = pd.concat([df, split_df], axis=1)
dfBB=df[df['channel'] == 'BHZ'] #sorting results for just broadband responses 
dfR=df[df['channel'] == 'EHZ']
dfRS=dfR[dfR['npts']==36001]
dfBB
```

```python
eventlist=dfRS.drop_duplicates(subset=['eventname'])
eventlist=eventlist['eventname']
df_final=pd.DataFrame()

for event in eventlist:
    dfBBevent=pd.DataFrame
    dfBBevent=dfBB[dfBB['eventname']==event]

    dfevent=pd.DataFrame()

    dfevent['eventname']=dfBBevent["eventname"]
    dfevent['BB Median Tacer Energy']=dfBBevent["ebbcorrmedtac"]
    dfevent['Station Name']=dfBBevent["station"]
    dfevent['Station Distance']=dfBBevent["dist_degrees"]
    dfevent['Event Azimuth']=dfBBevent["az"]
    dfevent['Event Rake']=dfBBevent["rake"]
    dfevent['Event Depth']=dfBBevent["edepth"]
    dfevent['Average BB Median Tacer Energy'] = dfevent["BB Median Tacer Energy"].mean()
    dfevent['BB Normalization']=np.log(dfevent['BB Median Tacer Energy']/dfevent['Average BB Median Tacer Energy'])

    df_final=pd.concat([df_final,dfevent],ignore_index=True)
    df_final.to_pickle("BBNormMeanErgs.pkl")
```

```python
dfBBB = pd.read_pickle('BBNormMeanErgs.pkl')

distance = dfBBB['Station Distance'].tolist()
normalization = dfBBB['BB Normalization']


fig1 = px.scatter(x=distance, y=normalization, title='All Events, BB Stations Normalized by BB Mean', labels={"x":"Distance", "y":"Log Normalization of Median Tacer Energy"})
fig1.update_yaxes(range = [-10,10])
fig1
```
